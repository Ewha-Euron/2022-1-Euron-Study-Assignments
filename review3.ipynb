```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
HR=pd.read_csv('HR_employee_data.csv')
```

```python
HR.head()
```

```python
HR.describe()
```

```python
HR.drop(['EmployeeCount', 'EmployeeNumber', 'Over18','StandardHours'], axis=1, inplace=True)
```

```python
HR.columns
```

```python
HR['Attrition'].value_counts()
```

```python
HR.loc[HR.Attrition == "No", "Attrition"] = 0

HR.loc[HR.Attrition == "Yes", "Attrition"] = 1
```

```python
HR.info()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in HR.columns:
    
    le = LabelEncoder()
    le=le.fit(HR[col])
    HR[col]= le.transform(HR[col])

y=HR['Attrition']
X=HR.drop('Attrition',axis=1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
pred=dt.predict(X_test)
print(accuracy_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train, y_train)
pred=rf.predict(X_test)
print(accuracy_score(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV

params={
    'n_estimators':[30,40,50,70,80,90,100],
    'max_depth':[6,8,10,12]
}
rf=RandomForestClassifier(random_state=42)
grid_cv=GridSearchCV(rf, param_grid=params)
grid_cv.fit(X_train, y_train)
print(grid_cv.best_params_)
```

```python
estimator=grid_cv.best_estimator_
pred=grid_cv.predict(X_test)
print(accuracy_score(y_test,pred))
```
